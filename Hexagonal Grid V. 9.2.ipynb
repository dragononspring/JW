```python
import numpy as np
import math
import matplotlib.pyplot as plt
import random
from csv import writer
from csv import reader
import time
from ipywidgets import interactive
from numpy import loadtxt

ninitial  = 3 #initial particle randomly placed on surface as seeds(set at center must be 3)
dn = 1 # number of droping each time 

xlength = 100 #"length of Y axis" where the lenght is sqrt 3 each
ylength = 58 #"length of X axis"
# actually available position of particle limits in x in  [0,ylength-1],[0,xlength-1]

nmax = int(xlength*ylength*2*0.1) # total number of droping, need to be int

nu = 10**13 # prefactor of Arrhenius Equation 

temperature = 300 # Tempearture of Arrhenius Eq.

nmolecules = nmax + 1 

kb = 8.617333262*10**-5

beta = 1/(temperature*kb)

BE = 1.5 # Energy of breaking a dimer

E1 = 0.3 #Ea values for 37, Diffusion Barrier

E3 = np.array([0,0.2,0.15,0.1,0.05,0.05,0]) #Barrier of forming bond with differnt number of neighbors

E2 = E3 + np.array([0,1,1.45,1.9,2.35,2.35,0])* BE #Barrier of Breaking bond with different number of neighbors

E4 = BE*0.7*np.array([0,1,1])

OneTimeCounter = 0 # Error check

Looplimit = 100000 #limit loops of diffuison between each drop

Limit = 100000000 #limit total loops for heating 

Ntolerance = 6 # Limiting numbers of neighbor one can bond with yet 0 is not working. eg. 1 is dimer; 2 is trimer + wire, etc. 

KDropping = 1 # rate of dropping, 1 means 1 cluster per sec.

def Initialization(ni):
    global z,n,t, PossibleMoves,PossibleAllNeighbors,AnimationMatrix,Loops,oneLoops
    z = np.empty((ni,3))
    AnimationMatrix = np.zeros((1,nmolecules,6),dtype=int)
    n = ni
    t=0
    i=0
    Loops = 0
    oneLoops = Loops
    #force the initial 3 at center
    z[0] = [0,int(ylength/2),int(xlength/2)]
    z[1] = [1,int(ylength/2),(int(xlength/2))]
    z[2] = [0,int(ylength/2),(int(xlength/2)+1)]
    #while i < ni:
        #ai = random.randint(0,1)
        #ri = random.randint(0,ylength-1)
        #ci = random.randint(0,xlength-1)
        #newmolecule = np.array([ai,ri,ci])
        #if not(newmolecule == z).all(axis=1).any():
            #z[i] = newmolecule
            #i += 1
    for iii in range(n):
        AnimationMatrix[0,iii] = [z[iii,0],z[iii,1],z[iii,2],n,Loops,t]
                
def MovesGenerator(mi):
    global z, PossibleMoves,AllPossibleNeighbors
    an,rn,cn = mi[0], mi[1], mi[2]
    PossibleMoves=np.array([[an,rn,cn+1],[1-an,rn+an-1,cn+an],[1-an,rn+an-1,cn+an-1],[an,rn,cn-1],[1-an,rn+an,cn+an-1],[1-an,rn+an,cn+an]])
    for i in range(6): #check if any position go to boundary, if so move to the other side
        if (PossibleMoves[i,1] >= ylength):
            PossibleMoves[i,1] += -ylength
        elif (PossibleMoves[i,1] < 0):
            PossibleMoves[i,1] += ylength
        if (PossibleMoves[i,2] >= xlength):
            PossibleMoves[i,2] += -xlength
        elif (PossibleMoves[i,2] < 0):
            PossibleMoves[i,2] += xlength
    return(PossibleMoves)   

def AllNeighborsGenerator(mi):
    global z, PossibleMoves,AllPossibleNeighbors
    an,rn,cn = mi[0], mi[1], mi[2]
    AllPossibleNeighbors = np.array([ [an,rn,cn+1],[1-an,rn+an-1,cn+an],[1-an,rn+an-1,cn+an-1],[an,rn,cn-1],[1-an,rn+an,cn+an-1],[1-an,rn+an,cn+an], [an,rn,cn+2], [1-an,rn+an-1,cn+an+1], [an,rn-1,cn+1],[an,rn-1,cn],[an,rn-1,cn-1], [1-an,rn+an-1,cn+an-2] ,[an,rn,cn-2],[1-an,rn+an,cn+an-2],[an,rn+1,cn-1],[an,rn+1,cn],[an,rn+1,cn+1],[1-an,rn+an,cn+an+1]   ])
    for i in range(18): #check if any position go to boundary, if so move to the other side
        if (AllPossibleNeighbors[i,1] >= ylength):
            AllPossibleNeighbors[i,1] += -ylength
        elif (AllPossibleNeighbors[i,1] < 0):
            AllPossibleNeighbors[i,1] += ylength
        if (AllPossibleNeighbors[i,2] >= xlength):
            AllPossibleNeighbors[i,2] += -xlength
        elif (AllPossibleNeighbors[i,2] < 0):
            AllPossibleNeighbors[i,2] += xlength    
    return(AllPossibleNeighbors)
                
def plotting(m):
    global plottingmatrix
    nparticles=m.shape[0]
    plottingmatrix = np.empty((nparticles,2))
    for i in range(nparticles):
        x = (m[i,0]/2 + m[i,2])
        y = np.sqrt(3) * (m[i,0]/2 + m[i,1])
        plottingmatrix[i] = [x,y]
    return(plottingmatrix)


def AnimationTracker(z):
    global AnimationMatrix, TempMatrix,AnimationMatrix, t, n, loops
    TempMatrix = np.zeros((1,nmolecules,6),dtype=int)
    for iii in range(n):
        TempMatrix[0,iii] = [z[iii,0],z[iii,1],z[iii,2],n,Loops,t]
    AnimationMatrix=np.append(AnimationMatrix,TempMatrix,axis=0)      

def TimePlot(r=0):
    global AnimationMatrix,AnimationPlot,InteractivePlot,n,ninitial,rrange
    PlottingRange=AnimationMatrix[r,0,3]
    AnimationPlot=np.empty((PlottingRange,2))
    for j in range(PlottingRange):
        AnimationPlot[j] = [ AnimationMatrix[r,j,0]/2 +  AnimationMatrix[r,j,2]  , np.sqrt(3)* ( AnimationMatrix[r,j,0]/2 +  AnimationMatrix[r,j,1]   )      ]
    plt.scatter(AnimationPlot[0:PlottingRange,0],AnimationPlot[0:PlottingRange,1], color = 'b')
    plt.title("Time Dependent")
    plt.xlim(-1,xlength+1)
    plt.ylim(-1,np.sqrt(3)*ylength+1)
    plt.figure(figsize=(15,15))
    rrange=n-ninitial
    InteractivePlot=interactive(TimePlot,r=(0,rrange))
    
    
    
def NeighborCheck(singlepart):
    global NeighborMatrix,z,NNMatrix,MoveNNMatrix,ToleranceNNMatrix,RotationNNMatrix
    MovesGenerator(z[singlepart])
    PossibleNeighbors = np.copy(PossibleMoves)
    PossibleNeighbors1 = PossibleNeighbors
    j = 0
    zcopy = np.copy(z)
    zcopy=np.delete(zcopy,(singlepart),axis=0)
    for iii in PossibleNeighbors:
        NewLocation = np.array(iii)
        if not(NewLocation == z).all(axis=1).any(): #If there is a no particle there, Value in NNMatrix is 0
            NNMatrix[singlepart,j] = 0
            MoveNNMatrix[singlepart,j] = 0
            RotationNNMatrix[singlepart,j] = 0
            MovesGenerator(NewLocation)
            PossibleNeighbors = np.copy(PossibleMoves)
            for q in PossibleNeighbors:
                SecondNNLocation = q
                if (SecondNNLocation == zcopy).all(axis=1).any():
                    MoveNNMatrix[singlepart,j] += 1
                    if(SecondNNLocation == PossibleNeighbors1).all(axis=1).any():
                        RotationNNMatrix[singlepart,j] += 1
        else: 
            NNMatrix[singlepart,j] = 1
            MoveNNMatrix[singlepart,j] = 0
            RotationNNMatrix[singlepart,j] = 0
        j+=1
    NNMatrix[singlepart,6] = NNMatrix[singlepart,0:6].sum() 
    ToleranceNNMatrix[singlepart,6] = NNMatrix[singlepart,6]

def TNeighborCheck(singlepart):
    global NeighborMatrix,z,NNMatrix,MoveNNMatrix,ToleranceNNMatrix,RotationNNMatrix
    MovesGenerator(z[singlepart])
    PossibleNeighbors = np.copy(PossibleMoves)
    PossibleNeighbors1 = PossibleNeighbors
    k = 0
    zcopy = np.copy(z)
    zcopy = np.delete(zcopy,(singlepart),axis=0)
    for kkk in PossibleNeighbors:#need to make the moving particle not effecting NNMatrix, but how to make this smaller?    
        SecondNNLocation = np.array(kkk)
        if (SecondNNLocation == zcopy).all(axis=1).any():
            Index = int(np.where((SecondNNLocation == z).all(1))[0])
            ToleranceNNMatrix[Index,6] += -1
    for jjj in PossibleNeighbors:
        NewLocation = np.array(jjj)
        if not(NewLocation == z).all(axis=1).any():#If there is a no particle there,
            MovesGenerator(NewLocation)
            PossibleNeighbors = np.copy(PossibleMoves)
            neighbortolerance = 1
            ToleranceNNMatrix[singlepart,k] = 1
            for r in PossibleNeighbors:
                SecondNNLocation = r
                if (SecondNNLocation == zcopy).all(axis=1).any():
                    Index = int(np.where((SecondNNLocation == z).all(1))[0])
                    if (ToleranceNNMatrix[Index,6] >= Ntolerance):
                        neighbortolerance = 0
            ToleranceNNMatrix[singlepart,k] = neighbortolerance
        else:
            ToleranceNNMatrix[singlepart,k] = 0
        k+=1
    for kkk in PossibleNeighbors1:#need to make the moving particle not effecting NNMatrix, but how to make this smaller?    
        SecondNNLocation = np.array(kkk)
        if (SecondNNLocation == zcopy).all(axis=1).any():
            Index = int(np.where((SecondNNLocation == z).all(1))[0])
            ToleranceNNMatrix[Index,6] += 1


            
def TotalNNCheck(npart):
    global NNMatrix,MoveNNMatrix,ToleranceNNMatrix,RotationNNMatrix
    NNMatrix = np.zeros((npart,7))
    MoveNNMatrix = np.zeros((npart,6))
    ToleranceNNMatrix = np.zeros((npart,7)) + 1
    RotationNNMatrix = np.zeros((npart,6))
    for l in range(npart):
        NeighborCheck(l)
    for l in range(npart):
        TNeighborCheck(l)        
    return (NNMatrix,MoveNNMatrix,ToleranceNNMatrix,RotationNNMatrix)

def SingleKCompute(singlepart): 
    global NNMatrix,MoveNNMatrix,KMatrix,temperature,kb,ToleranceNNMatrix,RotationNNMatrix
    for i in range(6):
        if ToleranceNNMatrix[singlepart,i] == 1: # NNMatrix[singlepart,i] == 0 included 
            KMatrix[singlepart,i] = nu * np.exp(-1/(temperature*kb)*( E1 * np.heaviside( 1 -  NNMatrix[singlepart,6] -int(MoveNNMatrix[singlepart,i])  ,0)  + E2[int(NNMatrix[singlepart,6])] - E4[int(RotationNNMatrix[singlepart,i])] + E3[int(MoveNNMatrix[singlepart,i])]) )
        else:
            KMatrix[singlepart,i] = 0
    KMatrix[singlepart,6] = np.sum(KMatrix[singlepart,0:6])
    return(KMatrix)


def TotalKCompute(npart):
    global NNMatrix,MoveNNMatrix,KMatrix,temperature,kb,ToleranceNNMatrix,RotationNNMatrix
    KMatrix = np.zeros((npart,7))
    for l in range(npart):
        SingleKCompute(l)
    return(KMatrix)

def NewParticle(nnumber):
    global z
    ii = 0
    while ii < nnumber:
        ai = random.randint(0,1)
        ri = random.randint(0,ylength-1)
        ci = random.randint(0,xlength-1)
        newmolecule = np.array([ai,ri,ci])
        if not(newmolecule == z).all(axis=1).any():
            z= np.append(z,[newmolecule],axis=0)
            ii+= 1
    return(z)

def MoveDecider(AdditionalN):
    global KMatrix, n,t,nmax,z, PossibleAllNeighbors, KMatrix, NNMatrix,MoveNNMatrix, dn,Loops, OneTimeCounter,ToleranceNNMatrix,RotationNNMatrix,Looplimit,oneLoops
    KMoving = np.sum(KMatrix[:,6])#no need to check off grid as boundary repeat
    if AdditionalN == 1:
        nmolecules = nmax + 1
    else:
        nmolecules = nmax
    if n < nmolecules:
        KTot = KMoving + KDropping
    else:
        KTot = KMoving
    r1,r2=random.random(),random.random()
    if KTot != 0:
        t += -1*math.log(r1)/KTot
    else: t+= 0
    KMod = r2 * KTot
    if KTot == 0 and OneTimeCounter == 0:
        np.savetxt('KTotalZero.txt', z, delimiter='\t', fmt='%f')
        OneTimeCounter += 1
    if KTot >0 :
        if (KMod <=  KMoving) and (Loops < (oneLoops + Looplimit)):#if moving
            mparticle = 0
            TempSum = KMatrix[mparticle,6]
            while TempSum < KMod:
                mparticle += 1
                TempSum += KMatrix[mparticle,6]
            TempSum += -KMatrix[mparticle,6]
            mm = 0
            TempSum += KMatrix[mparticle,mm]
            while TempSum < KMod:
                mm += 1
                TempSum += KMatrix[mparticle,mm]
            zold = np.copy(z[mparticle])#diffusion process where zold is origin position of the particle
            MovesGenerator(z[mparticle])
            z[mparticle] = PossibleMoves[mm]#search for that specific move and move particle to that position
            AllNeighborsGenerator(zold)#upgrade the NNMatrix, MoveNNMatrix and ToleranceNNMatrix
            Search= np.copy(AllPossibleNeighbors)
            AllNeighborsGenerator(z[mparticle])
            Search = np.append(Search,AllPossibleNeighbors, axis = 0)
            Search = np.unique(Search,axis=0) #Checks Over original Position, but I wanted to make the code shorter
            for jj in Search:
                if (jj == z).all(axis=1).any():
                    Index = int(np.where((jj == z).all(1))[0])
                    NeighborCheck(Index)
            for kk in Search:
                if (kk == z).all(axis=1).any():
                    Index = int(np.where((kk == z).all(1))[0])
                    TNeighborCheck(Index)
                    SingleKCompute(Index)

        else:#if not moving then drop more
            oneLoops = Loops
            if n+ dn >nmolecules:
                dn = nmolecules-n
            NewParticle(dn)
            n += dn
            TotalNNCheck(n)
            TotalKCompute(n)
            print('Total Particles is:',n, " At Code Time: ", t, " Loops:", Loops)
            AnimationTracker(z)
        return(n,z)


def zImport(Name):
    global z,n,KMatrix,t,nmolecules, PossibleAllNeighbors, KMatrix, NNMatrix,MoveNNMatrix, dn,Loops,ToleranceNNMatrix,RotationNNMatrix
    z = np.genfromtxt(Name+'.txt', dtype=int,encoding=None, skip_footer=0)
    n = z.shape[0]
    TotalNNCheck(n)
    TotalKCompute(n)

def Heating(Import,Name,Temper, Tmax, TStep, LoopMax):
    global KMatrix, n,t,nmolecules,z, PossibleAllNeighbors, KMatrix, NNMatrix,MoveNNMatrix, dn,Loops,temperature,ToleranceNNMatrix,RotationNNMatrix
    temperature = Temper
    t = 0
    if Import == 1:
        zImport(Name)
    Loops = 0
    TempLoops = 0
    LoopDiff = (((Tmax-temperature)/(TStep)) + 1)**(-1)  *LoopMax
    OneTimeCounter = 0
    while Loops < LoopMax:
        MoveDecider(0)
        Loops += 1
        if (Loops-TempLoops) > LoopDiff:
            temperature += TStep
            TempLoops = Loops
            TotalNNCheck(n)
            TotalKCompute(n)
            
def SingleHeating(Import,Name,Temper, tmax):
    global KMatrix, n,t,nmolecules,z, PossibleAllNeighbors, KMatrix, NNMatrix,MoveNNMatrix, dn,Loops,temperature,ToleranceNNMatrix,RotationNNMatrix
    temperature = Temper
    t = 0
    if Import == 1:
        zImport(Name)
    Loops = 0
    OneTimeCounter = 0
    while t <= tmax:
        MoveDecider(0)
        Loops += 1
        if Loops >= Limit:
            break

Initialization(ninitial)
TotalNNCheck(n)
TotalKCompute(n)




while n < nmolecules:
    MoveDecider(1)
    Loops+=1 
n+= -1
z = np.delete(z,n,axis=0)

plotting(z)




fig,(ax1)=plt.subplots(1,1,figsize=(10,10))
plt.ion()
ax1.set_title('time= ' + str(t) + 's')
ax1.set_aspect(1)
ax1.axes.set_xlim([-1,xlength+1])
ax1.axes.set_ylim([-1,np.sqrt(3)*ylength+1])
ax1.set_xlabel('nm')
ax1.set_ylabel('nm 300K')
ax1.scatter(plottingmatrix[0:n,0],plottingmatrix[0:n,1])
plt.show()

interactive(TimePlot,r=(0,n-ninitial))


np.savetxt('BaseCase300XY', plottingmatrix, delimiter='\t', fmt='%f')   # X is an array
np.savetxt('BaseCase300.txt', z, delimiter='\t', fmt='%f')   # X is an array

SingleHeating(1,'BaseCase300',450, 900)
plotting(z)

fig,(ax1)=plt.subplots(1,1,figsize=(10,10))
plt.ion()
ax1.set_title('time= ' + str(t) + 's')
ax1.set_aspect(1)
ax1.axes.set_xlim([-1,xlength+1])
ax1.axes.set_ylim([-1,np.sqrt(3)*ylength+1])
ax1.set_xlabel('nm 450K')
ax1.set_ylabel('nm')
ax1.scatter(plottingmatrix[0:n,0],plottingmatrix[0:n,1])
plt.show()
np.savetxt('XYAnnealingfinC01BE151145E302E4071015E107_450', plottingmatrix, delimiter='\t', fmt='%f')   # X is an array
np.savetxt('AnnealingfinC01BE151145E302E4071015E107_450.txt', z, delimiter='\t', fmt='%f')   # X is an array

SingleHeating(1,'BaseCase300',550, 900)
plotting(z)

fig,(ax1)=plt.subplots(1,1,figsize=(10,10))
plt.ion()
ax1.set_title('time= ' + str(t) + 's')
ax1.set_aspect(1)
ax1.axes.set_xlim([-1,xlength+1])
ax1.axes.set_ylim([-1,np.sqrt(3)*ylength+1])
ax1.set_xlabel('nm 550K')
ax1.set_ylabel('nm')
ax1.scatter(plottingmatrix[0:n,0],plottingmatrix[0:n,1])
plt.show()
np.savetxt('XYAnnealingfinC01BE151145E302E4071015E107_550', plottingmatrix, delimiter='\t', fmt='%f')   # X is an array
np.savetxt('AnnealingfinC01BE151145E302E4071015E107_550.txt', z, delimiter='\t', fmt='%f')   # X is an array

SingleHeating(1,'BaseCase300',700, 900)
plotting(z)

fig,(ax1)=plt.subplots(1,1,figsize=(10,10))
plt.ion()
ax1.set_title('time= ' + str(t) + 's')
ax1.set_aspect(1)
ax1.axes.set_xlim([-1,xlength+1])
ax1.axes.set_ylim([-1,np.sqrt(3)*ylength+1])
ax1.set_xlabel('nm 700K')
ax1.set_ylabel('nm')
ax1.scatter(plottingmatrix[0:n,0],plottingmatrix[0:n,1])
plt.show()
np.savetxt('XYAnnealingfinC01BE151145E302E4071015E107_700', plottingmatrix, delimiter='\t', fmt='%f')   # X is an array
np.savetxt('AnnealingfinC01BE151145E302E4071015E107_700.txt', z, delimiter='\t', fmt='%f')   # X is an array
```

```python
KTotalZerodef Heating(Import,Name,Temper, Tmax, TStep, LoopMax):
    global KMatrix, n,t,nmolecules,z, PossibleAllNeighbors, KMatrix, NNMatrix,MoveNNMatrix, dn,Loops,temperature
    temperature = Temper
    if Import == 1:
        zImport(Name)
    TotalNNCheck(n)
    TotalKCompute(n)
    Loops = 0
    TempLoops = 0
    LoopDiff = (np.abs((Tmax-temperature)/(TStep)) + 1)**(-1)  *LoopMax
    while Loops < LoopMax:
        MoveDecider(0)
        Loops += 1
        if (Loops-TempLoops) > LoopDiff:
            temperature += TStep
            TempLoops = Loops
```
